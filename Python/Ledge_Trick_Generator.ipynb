# Python functions for generating and displaying skate ledge trick combos

1. ledge_combo - generates each individual combo
2. display_combo - prints trick-in, trick-out, and double ledge combo given a skater's tricks in each stance

```python
#return a list of the intersection of two lists
def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3
```

```python
ledge = ['front 50-50' , 'front 5-0', 'back 50-50', 'back 5-0', 'back noseslide', 'front noseslide', 'front nosegrind', 'back nosegrind',
         'front lip', 'front smith', 'front tail', 'back lip', 'back smith', 'back crook', 'front suski', 'back suski', 'front crook', 'front blunt',
         'back blunt', 'back tail']

can_180_into = ['front 50-50', 'front 5-0', 'back 50-50', 'back 5-0', 'front nosegrind', 'back nosegrind', 'front crook', 'back crook', 'front smith', 'back smith']

exit_either = ['front 50-50', 'back 50-50']
exit_nollie = ['front nosegrind', 'back nosegrind', 'back noseslide', 'front noselide', 'front crook', 'back crook', 'back noseblunt', 'front noseblunt']
exit_regular = ['front tail', 'back tail', 'front 5-0', 'back 5-0', 'back suski', 'front suski','front smith', 'back smith', 'front blunt', 'back blunt', 'front lip', 'back lip']
```

```python
michael_regular = ['ollie','pop shuv','front 180', 'back 180', 'front shuv', 'kickflip', 'heelflip','backside flip','treflip','back bigspin']
michael_nollie = ['nollie', 'nollie shuv', 'nollie back 180','nollie front shuv']
michael_fakie = ['fakie ollie', 'fakie pop shuv', 'fakie front shuv', 'half cab', 'full cab', 'fakie front 180', 'fakie flip', 'fakie heel','fakie bigspin', 'fakie front bigspin']
michael_switch = ['switch ollie','switch front 180', 'switch pop shuv']
michael_ledge = ['front 50-50', 'front 5-0', 'back 50-50', 'back 5-0', 'front nosegrind', 'front tail', 'front lip']
```

```python
#trick in, trick out are booleans referring to whether or not we want a trick in and a trick out
#for example, trick_in = true, trick_out = false will give a combo with some trick in to manny but no trick out

#my_regular_tricks is a list of the skater's regular tricks, my_fakie_tricks is a list of their fakie tricks,  etc.
#my_ledge is a list of the skater's ledge tricks


def ledge_combo(trick_in, trick_out, my_regular_tricks, my_fakie_tricks, my_nollie_tricks, my_switch_tricks, my_grinds):
  #takes care of the case in which user wants a trick into the manual

  #check to see if the rider has any fakie or switch tricks. If they don't, the trick onto the ledge must land in regular stance so
  #they can get off the ledge
  if len(my_fakie_tricks) == 0 and len(my_switch_tricks) == 0:
    must_land_regular = True
  else:
    must_land_regular = False

  if trick_in == True:
    onto_ledge = random.choice(my_regular_tricks)
    stance_entering_ledge = exit_stance[onto_ledge]
    while must_land_regular ==  True and stance_entering_ledge == 'fakie':
      #if they can't do fakie/switch manuals and are landing in that stance we must rechoose the trick in at random
      onto_ledge = random.choice(my_regular_tricks)
      stance_entering_ledge = exit_stance[onto_ledge]

  else:
    stance_entering_ledge = 'regular'
    onto_ledge = 'ollie'

  #choose the manual trick
  if stance_entering_ledge == 'regular':
    ledge_trick = random.choice(my_grinds)
    
  else:
    #must choose a ledge trick that can be 180'd into
    #find the intersect between my_grinds and the tricks that can be 180'd into and choose a random one of these

    grind_choices = intersection(my_grinds, can_180_into)
    selected = random.choice(grind_choices)
    selected = selected.split()[1]
    ledge_trick = 'fakie ' +  selected

  #choose the trick out
  if trick_out == True:
    #case 1: enter ledge in regular and grinding/sliding on both trucks
    if stance_entering_ledge == 'regular' and ledge_trick in exit_either:
      #if they have nollie tricks, choose a random stance between nollie and regular and choose a random trick in the chosen stance
      if len(my_nollie_tricks) > 0:
        stance_out = random.choice([my_regular_tricks,my_nollie_tricks])
      #else choose regular stance for  trick out
      else:
        stance_out = my_regular_tricks

      trick_out = random.choice(stance_out)
      while trick_out == 'ollie':
        trick_out = random.choice(my_regular_tricks)
    #case 2: enter ledge in regular and grinding/sliding on the nose
    elif stance_entering_ledge == 'regular' and ledge_trick in exit_nollie:
      trick_out = random.choice(my_nollie_tricks)
      while trick_out == 'nollie':
        trick_out = random.choice(my_nollie_tricks)
    #case 3: enter ledge in regular and grinding/sliding on the tail
    elif stance_entering_ledge == 'regular' and ledge_trick in exit_regular:
      trick_out = random.choice(my_regular_tricks)
      while trick_out == 'ollie':
        trick_out = random.choice(my_regular_tricks)
    #case 4: enter ledge in fakie and grinding on both trucks
    elif stance_entering_ledge == 'fakie' and ledge_trick in exit_either:
      #see if they have any switch tricks that could be chosen to get off the ledge
      if len(my_switch_tricks) > 0:
        stance_out = random.choice([my_fakie_tricks,my_switch_tricks])
      #else choose fakie trick out
      else:
        stance_out = my_fakie_tricks
      trick_out = random.choice(stance_out)
    #case 5: enter ledge in fakie and grinding/sliding on the nose (e.g. back 180 nosegrind)
    elif stance_entering_ledge == 'fakie' and ledge_trick in exit_nollie:
      trick_out = random.choice(my_switch_tricks)
      while trick_out == 'switch ollie':
        trick_out = random.choice(my_switch_tricks)
    #case 6: enter ledge in fakie and grinding/sliding on the tail (e.g. back 180, fakie 5-0)
    else:
      trick_out = random.choice(my_fakie_tricks)
      while trick_out == 'fakie ollie':
        trick_out = random.choice(my_fakie_tricks)

  else:
    trick_out = ''

  if trick_out == '':
    print(onto_ledge, ',', ledge_trick)
  else:
     print(onto_ledge, ',', ledge_trick, ',', trick_out)
```

```python
def display_combo(my_regular_tricks, my_fakie_tricks, my_nollie_tricks, my_switch_tricks, my_grinds):
  print('Combo 1: ')
  ledge_combo(True,False, my_regular_tricks, my_fakie_tricks, my_nollie_tricks, my_switch_tricks, my_grinds)
  print('Combo 2: ')
  ledge_combo(False,True, my_regular_tricks, my_fakie_tricks, my_nollie_tricks, my_switch_tricks, my_grinds)
  print('Combo 3: ')
  ledge_combo(True,True, my_regular_tricks, my_fakie_tricks, my_nollie_tricks, my_switch_tricks, my_grinds)
```

```python
#example of calling the  display combo function and what it will yield:
display_combo(michael_regular, michael_fakie, michael_nollie, michael_switch, michael_ledge)
```
